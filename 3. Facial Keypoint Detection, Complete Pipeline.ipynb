1. Detect all the faces in an image using a face detector (we'll be using a Haar Cascade detector in this notebook).
2. Pre-process those face images so that they are grayscale, and transformed to a Tensor of the input size that your net expects. This step will be similar to the `data_transform` you created and applied in Notebook 2, whose job was tp rescale, normalize, and turn any iimage into a Tensor to be accepted as input to your CNN.

After each face has been appropriately converted into an input Tensor for your network to see as input, you can apply your `net` to each face. The ouput should be the predicted the facial keypoints.
```python
def show_all_keypoints(image, predicted_key_pts, gt_pts=None):
    """Show image with predicted keypoints"""
    # image is grayscale
    plt.imshow(image, cmap='gray')
    plt.scatter(predicted_key_pts[:, 0], predicted_key_pts[:, 1], s=20, marker='.', c='m')
    # plot ground truth points as green pts
    if gt_pts is not None:
        plt.scatter(gt_pts[:, 0], gt_pts[:, 1], s=20, marker='.', c='g')
```

```python
# visualize the output
# by default this shows a batch of 10 images
def visualize_output(test_images, test_outputs, gt_pts=None, batch_size=1):

    for i in range(batch_size):
        plt.figure(figsize=(20,10))
        ax = plt.subplot(1, batch_size, i+1)

        # un-transform the image data
        image = test_images[i].data   # get the image from it's Variable wrapper
        image = image.numpy()   # convert to numpy array from a Tensor
        image = np.transpose(image, (1, 2, 0))   # transpose to go from torch to numpy image

        # un-transform the predicted key_pts data
        predicted_key_pts = test_outputs[i].data
        predicted_key_pts = predicted_key_pts.numpy()
        # undo normalization of keypoints  
        predicted_key_pts = predicted_key_pts*50.0+100
        
        # plot ground truth points for comparison, if they exist
        ground_truth_pts = None
        if gt_pts is not None:
            ground_truth_pts = gt_pts[i]         
            ground_truth_pts = ground_truth_pts*50.0+100
        
        # call show_all_keypoints
        show_all_keypoints(np.squeeze(image), predicted_key_pts, ground_truth_pts)
            
        plt.axis('off')

    plt.show()
```

```python
image_copy = np.copy(image)

PADDING = 55
shiftUp = -10
shiftLeft = 5
images, keypoints = [], []

# loop over the detected faces from your haar cascade
for (x,y,w,h) in faces:
    # Calculate the padding
    final_hw = max(h,w)
    #PADDING = int(final_hw/3)
    
    # Select the region of interest that is the face in the image 
    #roi = image_copy[y:y+h, x:x+w]
    roi = image_copy[y-PADDING-shiftUp:y+h+PADDING-shiftUp, x-PADDING-shiftLeft:x+w+PADDING-shiftLeft]
    
    # if image has an alpha color channel, get rid of it
    if(roi.shape[2] == 4):
        roi = roi[:,:,0:3]
            
    ## TODO: Rescale the detected face to be the expected square size for your CNN (224x224, suggested)
    h, w = roi.shape[:2]
    if h > w:
        new_h, new_w = h * 224/w, 224
    else:
        new_h, new_w = 224, w * 224/h
    new_h, new_w = int(new_h), int(new_w)
    face_copy = cv2.resize(roi, (new_w, new_h))
    if new_h > new_w:
        top = (new_h-new_w)/2
        top = int(top)
        left = 0
    else:
        top = 0
        left = (new_w-new_h)/2
        left = int(left)
    face_copy = face_copy[top: top + 224,left: left + 224]
    
    ## TODO: Convert the face region from RGB to grayscale
    # convert image to grayscale
    face_copy = cv2.cvtColor(face_copy, cv2.COLOR_RGB2GRAY)
    
    ## TODO: Normalize the grayscale image so that its color range falls in [0,1] instead of [0,255]
    face_copy = face_copy/255.0
    
    ## TODO: Reshape the numpy image shape (H x W x C) into a torch image shape (C x H x W)
    # if image has no grayscale color channel, add one
    if(len(face_copy.shape) == 2):
        # add that third color dim
        face_copy = face_copy.reshape(face_copy.shape[0], face_copy.shape[1], 1)
    # swap color axis because
    # numpy image: H x W x C
    # torch image: C X H X W
    #print(face_copy.shape)
    face_copy = face_copy.transpose((2, 0, 1));
    # Match the convolution dimensions
    face_copy = np.expand_dims(face_copy, axis=0)
    #print(face_copy.shape)
    face_copy = torch.from_numpy(face_copy)
    
    ## TODO: Make facial keypoint predictions using your loaded, trained network 
    # convert images to FloatTensors
    face_copy = face_copy.type(torch.FloatTensor)
       
    # forward pass to get net output
    output_pts = net(face_copy)
    # reshape to batch_size x 68 x 2 pts
    test_outputs = output_pts.view(output_pts.size()[0], 68, -1)
    
    ## TODO: Display each detected face and the corresponding keypoints        
    visualize_output(face_copy, test_outputs)
```
